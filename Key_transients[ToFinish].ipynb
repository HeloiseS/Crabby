### Now need the contextual features

ra 	dec 	rb_pix 	z 	photoz 	ebv_sfd 	log10_sep_arcsec 	SN 	NT 	ORPHAN 	CV 	UNCLEAR

```python
from astropy.coordinates import SkyCoord
from dustmaps.sfd import SFDQuery

class ContextualFeatures():
    context_feature_columns = ['ra',
                               'dec',
                               'rb_pix',
                               'z',
                               'photoz',
                               'ebv_sfd',
                               'log10_sep_arcsec',
                               'SN',
                               'NT',
                               'ORPHAN',
                               'CV',
                               'UNCLEAR'
                              ]
    def __init__(self, atlas_json_data):
        self.data = atlas_json_data
        
        self._get_extinction()
        
    
        if len(self.data.data['sherlock_crossmatches']) != 0:
            self.context_features = [self.data.data['object']['ra'],
                                   self.data.data['object']['dec'],
                                   self.data.data['object']['rb_pix'],
                                   self.data.data['sherlock_crossmatches'][0]['z'],
                                   self.data.data['sherlock_crossmatches'][0]['photoz'],
                                   self.ebv,
                                   np.log10(self.data.data['sherlock_crossmatches'][0]['separationarcsec'])
                                   ]
        else:
            self.context_features = [self.data.data['object']['ra'],
                                   self.data.data['object']['dec'],
                                   self.data.data['object']['rb_pix'],
                                   np.nan,
                                   np.nan,
                                   self.ebv,
                                   np.nan
                                   ]
        self.context_features = self.context_features + self.data.make_sherlock_features()
            
    def _get_extinction(self):
        coords = SkyCoord(str(self.data.data['object']['ra']), 
                  str(self.data.data['object']['dec']), 
                  frame='icrs', unit='deg')
        sfd = SFDQuery()
        self.ebv = sfd(coords)
        
```

## check actual ranks

```python
def rank_function(REAL_PRED, GAL_PRED, fudge_factor=0.5, max_score=10):
    ranks = np.sqrt((1-REAL_PRED)**2 + fudge_factor*(GAL_PRED)**2)
    max_distance = np.sqrt(fudge_factor**2 + 1)
    return (max_distance - ranks) * max_score / max_distance
```
