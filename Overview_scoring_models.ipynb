# Overview of training and permutation importance

The goal of this notebook is to show a demo of how we're training the models before we abstract everything away into a couple of python scripts that will run in a loop. 

```python
############################################################
#          _____ ____________             __  ______        
#         / ___// ____/_  __/            / / / / __ \       
#         \__ \/ __/   / /              / / / / /_/ /       
#        ___/ / /___  / /              / /_/ / ____/        
#       /____/_____/ /_/               \____/_/             
#
###########################################################

%load_ext autoreload
%autoreload 2

##########
# IMPORTS
##########

from atlasvras.utils.prettify import vra_colors, label_to_color
import matplotlib.pyplot as plt
from matplotlib import gridspec
import pandas as pd 
from glob import glob
from joblib import load
from tqdm.notebook import tqdm
import numpy as np
from sklearn.ensemble import HistGradientBoostingClassifier
from sklearn.inspection import PartialDependenceDisplay
from sklearn.inspection import permutation_importance

#### Utility dictionaries

dict_type_to_preal = {'garbage': 0,
                      'pm': 0,
                      'galactic': 1,
                      'good': 1
                     }

dict_type_to_pgal = {'garbage': 0,
                      'pm': 1,
                      'galactic': 1,
                      'good': 0
                     }

def rank_function(REAL_PRED, GAL_PRED, fudge_factor=0.4, max_rank=10):
    """
    Function that calculates the VRA ranks from the real and galactic scores predicted by the models
    
    The "fudge_factor" is a scalar applied to the galactic axis and it is set to 0.5.
    The maximum rank is the maximum value we want the VRA ranks to take. 
    """
    ranks = np.sqrt((1-REAL_PRED)**2 + fudge_factor*(GAL_PRED)**2)
    max_distance = np.sqrt(fudge_factor**2 + 1)
    return (max_distance - ranks) * max_rank / max_distance



def recall_at_k(df_sorted, k, type_col='type'):
    # Get top K rows
    df_top_k = df_sorted.head(k)
    
    # Calculate precision and recall
    recall = len(df_top_k[df_top_k[type_col] == 'good']) / len(df_sorted[df_sorted[type_col] == 'good'])
    
    return recall


#  .+"+.+"+.+"+.+"+.+"+.+"+.+"+.+"+.+"+.+"+.+"+.+"+. 
# (        ____                                     )
#  )      / __/__ ____  ______ __                  ( 
# (      / _// _ `/ _ \/ __/ // /                   )
#  )    /_/  \_,_/_//_/\__/\_, /__  __     __      ( 
# (                       /___/ _ \/ /__  / /_      )
#  )                         / ___/ / _ \/ __/     ( 
# (                         /_/  /_/\___/\__/       )
#  )                                               ( 
# (                                                 )
#  "+.+"+.+"+.+"+.+"+.+"+.+"+.+"+.+"+.+"+.+"+.+"+.+" 

def plot_score_space(REAL_PRED, GAL_PRED, Y, TITLE = "SCORE SPACE [test]"):
    mask_garbage = Y.type == 'garbage'
    mask_pm = Y.type == 'pm'
    mask_good = Y.type == 'good'
    mask_galactic = Y.type == 'galactic'
    
    ###################

    fig = plt.figure(facecolor='black', figsize=(12, 8))
    gs = gridspec.GridSpec(3, 3, 
                           width_ratios=[1, 4, 1], 
                           height_ratios=[1, 4, 1], 
                           wspace=0.05, hspace=0.05)


    # Main 2D plot
    ax_main = plt.subplot(gs[1, 1])
    ax_main.scatter(REAL_PRED[mask_garbage], GAL_PRED[mask_garbage], 
               color=vra_colors['red'], 
                alpha=0.5,
                s=30, zorder=0.1)
    ax_main.scatter(REAL_PRED[mask_pm], GAL_PRED[mask_pm], 
               color=vra_colors['orange'], 
                alpha=0.5,
                s=30, zorder=0.5)

    ax_main.scatter(REAL_PRED[mask_galactic], GAL_PRED[mask_galactic], 
               color=vra_colors['yellow'], 
                alpha=0.9,
                s=30, zorder=10)

    ax_main.scatter(REAL_PRED[mask_good], GAL_PRED[mask_good], 
               color=vra_colors['blue'], 
                alpha=0.9,
                s=30, zorder=1000)

    ax_main.set_ylabel('Galactic Score')
    ax_main.set_xlabel('Real Score')

    # Top 1D marginalized distribution
    ax_top = plt.subplot(gs[0, 1], sharex=ax_main)
    __ = ax_top.hist(REAL_PRED[mask_garbage], bins=50, 
                     color=vra_colors['red'], 
                     edgecolor=vra_colors['red'],
                     alpha=0.7, 
                 density=True,
                     label='Garbage'
               )

    __ = ax_top.hist(REAL_PRED[mask_pm], bins=50, 
                     color=vra_colors['orange'], 
                     edgecolor=vra_colors['orange'],
                     alpha=0.7,
                 density=True,
                     label='PM'
               )

    __ = ax_top.hist(REAL_PRED[mask_galactic], bins=50, 
                     color=vra_colors['yellow'], 
                     edgecolor=vra_colors['yellow'],
                     alpha=0.6,
                 density=True,
                     label='Galactic'
               )

    __ = ax_top.hist(REAL_PRED[mask_good], bins=50, 
                     color=vra_colors['blue'], 
                     edgecolor=vra_colors['blue'],
                     alpha=0.6,
                 density=True,
                     label='Good'
               )

    # Right 1D marginalized distribution
    ax_right = plt.subplot(gs[1, 2], sharey=ax_main)



    __ = ax_right.hist(GAL_PRED[mask_garbage], bins=np.arange(0,1,0.01), orientation='horizontal',
                     color=vra_colors['red'], 
                     edgecolor=vra_colors['red'],
                     alpha=0.7, 
                 density=True
               )

    __ = ax_right.hist(GAL_PRED[mask_pm], bins=np.arange(0,1,0.01),  orientation='horizontal',
                     color=vra_colors['orange'], 
                     edgecolor=vra_colors['orange'],
                     alpha=0.7,
                 density=True
               )

    __ = ax_right.hist(GAL_PRED[mask_galactic], bins=np.arange(0,1,0.01),  orientation='horizontal',
                     color=vra_colors['yellow'], 
                     edgecolor=vra_colors['yellow'],
                     alpha=0.5,
                 density=True
               )

    __ = ax_right.hist(GAL_PRED[mask_good], bins=np.arange(0,1,0.01),  orientation='horizontal',
                     color=vra_colors['blue'], 
                     edgecolor=vra_colors['blue'],
                     alpha=0.5,
                 density=True
               )

    #### TITLE
    ax_top.set_title(TITLE, loc='right', fontsize=20)

    ax_top.legend(bbox_to_anchor=[1.01, 1.0, 0.0, 0.0])
##################################################################

############## PLOT STYLE ################
# Comment out if you don't have the file
plt.style.use('vra')
##########################################
```

# 1. Day 1 Models (all features)


## 1.1 Load the data

```python
RELATIVE_PATH = './data/features_and_labels_csv/day1/'
X_test_unbalanced = pd.read_csv(f'{RELATIVE_PATH}X_test_unbalanced_randomsplit.csv', index_col=0)
y_test_unbalanced = pd.read_csv(f'{RELATIVE_PATH}y_test_unbalanced_randomsplit.csv', index_col=0)

X_train = pd.read_csv(f'{RELATIVE_PATH }X_train_randomsplit.csv', index_col = 0)
y_train = pd.read_csv(f'{RELATIVE_PATH }y_train_randomsplit.csv', index_col = 0)

y_train_real = y_train.map(lambda x: dict_type_to_preal[x])
y_train_gal = y_train.map(lambda x: dict_type_to_pgal[x])

y_test_real = y_test_unbalanced.map(lambda x: dict_type_to_preal[x])
y_test_gal = y_test_unbalanced.map(lambda x: dict_type_to_pgal[x])
```

## 1.2 Training the Real and the Galactic scorers

The hyperparameters are chosen to be the ones that maximise our AuRaK metric (see the hyperparameter_tuning directory and notebook herein).

```python
real_scorer = HistGradientBoostingClassifier(learning_rate=0.1,
                                             l2_regularization=10,
                                             class_weight='balanced',
                                             random_state=42, 
                                             #categorical_features="from_dtype"
                                    )
gal_scorer = HistGradientBoostingClassifier(learning_rate=0.2,
                                            l2_regularization=10,
                                            class_weight='balanced',
                                             random_state=42, 
                                             #categorical_features="from_dtype"
                                    )

real_scorer.fit(X_train, y_train_real)
gal_scorer.fit(X_train, y_train_gal)
```

## 1.3 Predict the scores for the test sample and plot

The `.predict_proba()` method returns two values for each sample: 
* [Probability of False, Probability of True]

The "real" and "galactic" scores are therefore the second item in those lists (or item number 1 since python starts counting at 0). 

```python
real_scorer.predict_proba(X_test_unbalanced)
```

```python
# By transposing we can grab all the Real and Galactic scores in one go in index 1
real_pred_TEST = real_scorer.predict_proba(X_test_unbalanced).T[1]
gal_pred_TEST = gal_scorer.predict_proba(X_test_unbalanced).T[1]
```

```python
REAL_PRED = real_pred_TEST
GAL_PRED = gal_pred_TEST
Y = y_test_unbalanced

plot_score_space(REAL_PRED, GAL_PRED, Y)
```

```python
RANKS = rank_function(REAL_PRED, GAL_PRED)
df_ranks = pd.DataFrame(np.vstack((RANKS, y_test_unbalanced.values.T[0])).T, 
                 columns = ['rank', 'type'], index=y_test_unbalanced.index)

day1_ranks = df_ranks.sort_values('rank', ascending=False)
day1_ranks['k']=np.arange(1,df_ranks.shape[0]+1)
```

```python
recall_at_k_list_DAY1 = []

for k in np.arange(1, day1_ranks.shape[0]+1):
    _recall = recall_at_k(day1_ranks, k)
    recall_at_k_list_DAY1.append(_recall)
```

```python
fig, ax = plt.subplots(facecolor='black', figsize=(5,5))
ax.set_facecolor('black')
x=np.linspace(0,1, len(recall_at_k_list_DAY1))

ax.plot(x, recall_at_k_list_DAY1,  lw=1, alpha=1)

plt.ylim(0.5,1)
plt.xlim(0,0.6)
plt.title('Recall at Rank K [Day 1 - All features]')
```

## 1.4 Permutation Importance

```python
permimp_real = permutation_importance(real_scorer, 
                                      X_train, 
                                      y_train_real, 
                                      n_repeats=10, 
                                      random_state=42, 
                                      n_jobs=2
                                     )

permimp_gal = permutation_importance(gal_scorer, 
                                      X_train, 
                                      y_train_gal, 
                                      n_repeats=10, 
                                      random_state=42, 
                                      n_jobs=2
                                     )
```

```python
RESULT = permimp_real

sorted_importances_idx = RESULT.importances_mean.argsort()
importances = pd.DataFrame(
    RESULT.importances[sorted_importances_idx].T,
    columns=X_train.columns[sorted_importances_idx],
)
ax = importances.plot.box(vert=False, whis=10)
ax.set_title("Permutation Importances REAL SCORES")
ax.axvline(x=0, color="k", linestyle="--")
ax.set_xlabel("Decrease in accuracy score")
ax.figure.tight_layout()
```

```python
RESULT = permimp_gal

sorted_importances_idx = RESULT.importances_mean.argsort()
importances = pd.DataFrame(
    RESULT.importances[sorted_importances_idx].T,
    columns=X_train.columns[sorted_importances_idx],
)
ax = importances.plot.box(vert=False, whis=10)
ax.set_title("Permutation Importances GAL SCORES")
ax.axvline(x=0, color="k", linestyle="--")
ax.set_xlabel("Decrease in accuracy score")
ax.figure.tight_layout()
```
