# Summary Plots

This notebook presents some general properties of the data for the Crabby data set of the ATLAS VRA. 

* Alert type distribution
* Eyeballer decisions latency
* Recall at rank K if we only use the already existing RB score 

```python
%load_ext autoreload
%autoreload 2

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from atlasvras.utils.prettify import vra_colors

# note if you don't have this file style use your own and the plots below will just look different
plt.style.use('vra')
```

```python
vra_last_entries = pd.read_csv('./clean_data_csv/vra_last_entry_withmjd.csv', index_col=0)
detections = pd.read_csv('./clean_data_csv/detections_data_set.csv', index_col=0)
non_detections = pd.read_csv('./clean_data_csv/non_detections_100days.csv', index_col=0)
context = pd.read_csv('./clean_data_csv/contextual_info_data_set.csv', index_col=0)
```

```python
context=context[~context['type'].isna()]
vra_last_entries=vra_last_entries[~vra_last_entries['type'].isna()]
detections=detections[~detections['type'].isna()]
non_detections=non_detections[~non_detections['type'].isna()]
```

# 1. Alert type distribution

```python
## Calculate the fraction of each type of alert
frac_garbage = vra_last_entries[(vra_last_entries.preal == 0.0) 
                                  & (vra_last_entries.pgal != 1.0)].shape[0]/vra_last_entries.shape[0]

frac_pm = vra_last_entries[(vra_last_entries.preal == 0.0) 
                             & (vra_last_entries.pgal == 1.0)].shape[0]/vra_last_entries.shape[0]

frac_gal = vra_last_entries[(vra_last_entries.preal == 1.0) 
                             & (vra_last_entries.pgal == 1.0)].shape[0]/vra_last_entries.shape[0]

frac_good = vra_last_entries[(vra_last_entries.preal == 1.0) 
                             & (vra_last_entries.pgal != 1.0)].shape[0]/vra_last_entries.shape[0]

frac_poss = vra_last_entries[(vra_last_entries.preal == 0.5)].shape[0]/vra_last_entries.shape[0]

# just to check the totla fraction fromes to 1
frac_garbage + frac_pm + frac_gal + frac_good + frac_poss
```

```python
plt.figure(facecolor='black', figsize=(5,5))

labels = ['Garbage', 'PM Star', 'Galactic', 'Good']
sizes = [frac_garbage, frac_pm, frac_gal, frac_good]  # values for each section

# Create a pie chart
plt.pie(sizes, 
        labels=labels, 
        autopct='%1.1f%%',
        startangle=-15,
        colors=[vra_colors['red'], 
                vra_colors['orange'],
                vra_colors['yellow'],
                vra_colors['blue'], 
                ],
        textprops={'color':"white",'fontname': 'Lato'}
       )

# Equal aspect ratio ensures that pie is drawn as a circle.
plt.axis('equal')
plt.title(f'HUMAN LABELING\nSUMMARY (N={vra_last_entries.shape[0]})', fontsize=15, loc='left', color='w')
plt.savefig("./figures/pie_chart.png", bbox_inches='tight')
```

## 2. When are decisions made by eyeballers?

```python
# Split the data frame into one for each alert type as it makes plotting easier
vra_garbage = vra_last_entries[(vra_last_entries.preal==0.0)
                   & (vra_last_entries.pgal!=1.0)
                  ]

vra_pm = vra_last_entries[(vra_last_entries.preal==0.0)
                   & (vra_last_entries.pgal==1.0)
                  ]

vra_good = vra_last_entries[(vra_last_entries.preal==1.0) 
                              & (vra_last_entries.pgal==0.0)]

vra_gal = vra_last_entries[(vra_last_entries.preal==1.0)
                   & (vra_last_entries.pgal==1.0)
                  ]
```

```python
df_list = [vra_garbage, vra_pm, vra_good, vra_gal]
label_list = ['Garbage', 'PM Star', 'Good', 'Galactic']
colors=[vra_colors['red'], vra_colors['orange'], vra_colors['blue'], vra_colors['yellow']]

# Create a figure with a grid of subplots
fig, axes = plt.subplots(2, 2, 
                         figsize=(12, 8), 
                         constrained_layout=True, 
                         facecolor='black')

# Iterate over each subplot and plot the histogram for each dataframe
for i, ax in enumerate(axes.flatten()):
    # Adjust this line according to your dataframe columns
    hist_data = ax.hist(df_list[i]['ndays_to_decision'], 
                        bins=np.arange(0, 5, 0.041), 
                        color=colors[i])
    
    max_value = hist_data[0].max()
    for j, x in enumerate(np.arange(0.25, 2,0.25)):
        ax.axvline(x, ls='--', color='w', lw=1, alpha=0.7)
        ax.text(x - 0.14, max_value * 1.06, f'+{6 * (j+1)}h', fontsize=7)

    
    ax.set_title(f'{label_list[i]}', loc='right', fontsize=18, color=colors[i])
    ax.set_facecolor('none')

# Add a common title for all subplots
fig.suptitle('When are decisions made by eyeballers?', fontsize=20, ha='right')
for ax in axes.flatten()[2:]:
    ax.set_xlabel('Days after entering the eyeball list', fontsize=15)
for ax in axes.flatten()[0::2]:
    ax.set_ylabel('N')
    
plt.savefig("./figures/when_decision_made.png", bbox_inches='tight')
```

## 3. Real Bogus Score Distribution and Recall at Rank K

These are benchmarks for the models we're going to create.

### 3.1 Distribution of RB sscores for good and garbage classes

```python
__ = plt.hist(context[context.type=='garbage'].rb_pix, 
              bins = np.arange(0.2,1.01, 0.01),
              color=vra_colors['red'], alpha=1, label='Garbage'
             )



__ = plt.hist(context[context.type=='good'].rb_pix, 
              bins = np.arange(0.2,1.01, 0.01),
              color=vra_colors['blue'], alpha=0.8, label='Good'
             )

plt.legend(loc=2)
plt.ylabel('N')
plt.xlabel('Real/Bogus Score')
plt.savefig("./figures/RBscore_distrib.png", bbox_inches='tight')
```

### 3.2 Calculating the Recall at Rank K suing only the RB score. 

```python
def order_df_and_add_rank(df):
    "Function to order the data frame by RB score"
    new_df = df.sort_values('rb_pix', ascending=False)
    new_df['k']=np.arange(1,df.shape[0]+1)
    return new_df
```

```python
context = order_df_and_add_rank(context)
```

```python
def recall_at_k(df_sorted, k, type_col='type'):
    "Calculates the recall for the top k rows "
    # Get top K rows
    df_top_k = df_sorted.head(k)

    recall = len(df_top_k[df_top_k[type_col] == 'good']) / len(df_sorted[df_sorted[type_col] == 'good'])
    
    return recall
```

Now we loop over each k row in our ordered dataframe and claulate the recall using the function above. 
We record the results in a list and then we'll plot it. 

```python
recall_at_k_list = []

for k in np.arange(1, context.shape[0]+1):
    _recall = recall_at_k(context, k)
    recall_at_k_list .append(_recall)
```

### 3.3 Area under the Recall at rank K 

We also calculate the area under the RaK curve we just calculated, as this is one of the benchmarks we are going to compared our models against. 

```python
x=np.linspace(0,1, len(recall_at_k_list))
AuRaK= round(sum(np.ones(len(recall_at_k_list)) * np.diff(x)[0]*np.array(recall_at_k_list)),3)
```

```python
print(f'AuRaK when ordering by RB score: {AuRaK}')
```

```python
fig, ax = plt.subplots(facecolor='black', figsize=(5,5))
ax.set_facecolor('black')

ax.plot(np.linspace(0,1, len(recall_at_k_list)), 
        recall_at_k_list, label=f'Ordered by Real/Bogus Score\nAuRak = {AuRaK}', color=vra_colors['white'], lw=1
       )



ax.set_ylabel('Recall Fraction')
ax.set_xlabel('Fractional K')
#ax.set_title("Recall at rank K", loc='right', fontsize=20)
ax.legend()

#plt.xlim([0,0.6])
#plt.ylim([0.5,1])
plt.axhline(0.95, lw=1.5, ls='--', color=vra_colors['blue'])
plt.text(0.3, 0.88, '95% good alerts recovered', fontsize=10, color=vra_colors['blue'])
plt.savefig('./figures/RatK_rbpix.png', bbox_inches='tight')
```
