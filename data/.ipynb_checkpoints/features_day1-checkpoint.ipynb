# Make features to train the Day 1 eyeballer models

In this notebook we take the clean data we extracted from the json files and create the features we're going to use to train the models that score the alerts when they first enter the eyeball list (Day 1).

### Overview of the features

There are 19 features we use for the Day 1 models, which we can split into a few broad categories:

* **Light curve long term history (last 100 days)**
    * Nnondet_std: Standard deviation of the number of non detectoin between each detections
    * Nnondet_mean: Mean of the number of non detectoin between each detections
    * magdet_std: Standard deviation of the magnitude of each historical detection
* **Light curve recent history (last 5 days)**
    * DET_Nsince_min5d: Number of detections
    * NON_Nsince_min5d: Number of non detections
* **Postional scatter recent history (last 5 days)**
    * log10_std_ra_min5d: Log10 of the standard deviation of the RA
    * log10_std_dec_min5d: Log10 of the standard deviation of the Dec
* **Contextual Information**
    * ra: Ra
    * dec: Dec
    * rb_pix: Real/bogus score from the CNN	
    * z: Spectroscopic redshift
    * photoz: Photometric redshift
    * ebv_sfd: E(B-V)
    * log10_sep_arcsec: Log10 of the separation in arcsec from a nearby source 
    * Boolean flags for the following sherlock features:
        * SN
        * NT 	
        * ORPHAN 	
        * CV 	
        * UNCLEAR

## 3. LightCurve Features

The light curve features try to capture the **history of detections in the 100 days prior to entering the eyeball list** as well as the **scatter on sky** of the recent detections (within 5 days w.r.t entering eyeball list). 

The following features are relevant to the long term history:
- `Nnondet_std`	
- `Nnondet_mean` 	
- `magdet_std` 	


`Nnondet` is the number of non detections in between detections in the history. We use the mean and standard deviation of this value as features of the history. `magdet` is the magnitude of each past detection, and we use the standard deviation as a feature. We do not use the mean as past iterations of the VRA showed that the mean of that value was not informative to the models. 

The following features look at the recent history:
- `DET_Nsince_min5d`: number of detections over 5 days before entering eyeball list
- `NON_Nsince_min5d`: number of non detections over 5 days before entering eyeball list	


The following features are relevant to the scatter on the sky:
- `log10_std_ra_min5d`: log10 of the standard deviation of the RA in the detections in the past 5 days
- `log10_std_dec_min5d`: log10 of the standard deviation of the Dec in the detections in the past 5 days


**Why 5 days?**: Because given the ATLAS cadence that usually gives us one or two aditional data points (on top of the data point of the alert) whether they be detections or non detections. That's usually the time window where we might have some detections that were a bit faint to pass the quality cuts of the ATLAS transient server ingest. Looking back a few days is useful to human eyeballers, therefore it will likely be informative to the model.


### 3.1 Make the Lightcurve (and crop)

Our **lightcurve** is the combination of our **detections + non_detections**. We crop it with respect to the `phase_init` which is **number of days since the alert entered the VRA**.

Since we are making the **day 1 models (a.k.a eyeballing on first entry in the eyeball list)** we only want the data  that was taken **before phase 0**.

```python
HI_CUT_OFF = 0 
LO_CUT_OFF = -100
```

```python
# ### CROP TO THE WINDOW OF TIME SPECIFIED ### #
detections = detections[(detections['phase_init'] < HI_CUT_OFF )
                                                    & (detections['phase_init'] >= LO_CUT_OFF)]
non_detections = non_detections[(non_detections['phase_init'] < HI_CUT_OFF )
                                                    & (non_detections['phase_init'] >= LO_CUT_OFF)]

# ### CONCATENATE TO LIGHTCURVE ### #
lightcurve = pd.concat([detections, non_detections]).sort_values(by='phase_init')
```

Order the dataframe so the data for each object is grouped and in increasing time order. 

```python
lightcurves = lightcurve.reset_index('ATLAS_ID').sort_values(['mjd', 'ATLAS_ID']).set_index('ATLAS_ID')
```

```python
lightcurves
```

### 3.2 Find the first (and last) detections in each lighcurve

To calculate the `Nnon_det`, the number of detections in between each historical detection, we first find the very first detection and crop the lightcurve at that position. 
That's because I want to measure the **mean** and **standard deviation** of the number of non detections between each detection. **I need detections to anchor my count** - if my lightcurve starts with non detections, I don't know if it's been 5 days 5 months or 5 years since something was detected at that location (or maybe there;s never been a detection at that position). 

**I can only start counting when I've found a detection, so we're going to find the first in each lightcurve and crop everything that happened before**. 

For some lightcurves that means we crop everything, and that's okay. If they have no lightcurve history that is information that our model can use. 


```python
# calculating the first and last MJDs requires a little more work as we have to group
# First we create pandas Series with the first and last MJDs for each object and ATLAS_ID as the index
first_det_mjds = lightcurves[lightcurves.det == True].groupby('ATLAS_ID').min().mjd
last_det_mjds = lightcurves[lightcurves.det == True].groupby('ATLAS_ID').max().mjd
```

```python
first_det_mjds
```

```python
last_det_mjds
```

We add the dates of the first and last detections as additional columns in the lightcurves table 

```python
# Then we add the columns by mapping the ATLAS_ID to the series
# right now can't remember why this isn't just a .join(). I think it's because that only works
# on pandas dataframes but worth checking.
lightcurves['first_mjd'] = lightcurves.index.map(first_det_mjds)
lightcurves['last_mjd'] = lightcurves.index.map(last_det_mjds)
```

```python
lightcurves
```

Now we crop the lightcurves starting from the first detection. 

```python
# Alright now we crop all the data before and including the first historical detection.
# If you're wondering why that's just how I'm defining these features.
cropped_lc = lightcurves[lightcurves['mjd'] > lightcurves['first_mjd']]
```

### 3.3 Locate the detections and count the non detections in between 

To find `Nnondet` we need to group our non detections  - we cna do that by adding a column at each row that states how many previous detections have been seen (starting at 0 because it's python. 

If it's not clear scroll down to the next table and paragraph of text.

```python
def _calculate_N_dets(group):
    """
    Calculates the number of cummulative number detections for a given object and puts it in a column
    This is a function because when handling a whole sample of objects with a dataframe with index ATLAS_ID
    We can do this operation without a loop by grouping and then using this function with the .apply() method

    I could probably have made a lambda function inside .apply() but this was more readable.
    """
    group['N_dets'] = group['det'].cumsum()
    return group
```

```python

# we now group by ATLAS_ID and apply the function that calculates the cumulative number of detections
grouped_crop = cropped_lc.groupby(cropped_lc.index,
                                  group_keys=False  # added this because pandas gave me a warning and told me to
                                  )
cropped_lc = grouped_crop.apply(_calculate_N_dets)
```

```python
cropped_lc.loc[1045053540725632500]
```

In this table we can see what the `N_dets` column does: each row has a counter and the counter only goes up with there is a detection. 
If we only look at non-detections, then the ones that should be groupped all have the same `N_dets`. 

**Now here is the python-fu that make Nnon_det happen**

* 1) Group by `ATLAS_ID` and `N_dets` -> You've broken up your lightcurve into segments split by each new detection. (and of course ATLAS\_ID). 
* 2) Focus on the boolean `det` column 
* 3) Do  `det == False` => You get 1 for each row that is a non detection
* 4) The sum of that mask is the total number of non detection in your group



```python
n_nondets = cropped_lc.groupby(['ATLAS_ID', 'N_dets'])['det'].apply(lambda x: (x == False).sum())
```

```python
n_nondets
```

### 3.4 Make our three 'long term' history features 

- `Nnondet_std`	
- `Nnondet_mean` 	
- `magdet_std` 	


```python
n_nondets_std = n_nondets.groupby('ATLAS_ID').std()
n_nondets_mean = n_nondets.groupby('ATLAS_ID').mean()
```

```python
mag_std = lightcurves[lightcurves.det == True].groupby('ATLAS_ID').mag.std()
```

### 3.5 Make the 'short term' history features 
Those associated with the 5 days preceeding entry into the eyeball list.
To do this we create the `dayN` column. 
Note that's not entierly necessary just for this but it's also a column used when making the features for the update model so we use it here too to define our '> -5 days' crops.

#### The `dayN` column
To our lightcurves we add the `dayN` column with is the integer number of days since the 'first' detections:

* The **zeroth day** is the day when the **observations that triggered entry in the eyeball list** were made.
Since these observations are made just before the object enters the eyeball list their `phase_init`values are NEGATIVE.
Our zeroth day therefore corresponds to phase_init [-1, 0]. 
* Our **day 1** corresponds to phase_init (0,1]


Consequently dayN is the integer part of phase_init for NEGATIVE phases.
For positive phases dayN = int(phase_init) + 1. **Without this correction day 0 would last  2 days.**

```python
def add_dayN_col(df: pd.DataFrame) -> pd.DataFrame:
    """
    Adds the dayN column to a dataframe (if it has a pre-existing phase_init column).
    dayN is NOT SIMPLY THE INTEGER PART OF phase_init. See Notes for more details.
    
    Parameters
    ----------
    df : pd.DataFrame
        Dataframe of the lightcurve data
        The dataframe must have a column called 'phase_init' which is the
        phase w.r.t the date when the alert entered the eyeball list, that is when
        it passed the quality checks and was deemed a real alert
        (see manual and Smith et. al. 2021 for more details).

    Returns
    -------
    pd.DataFrame
        The dataframe with the dayN column added

    Raises  
    ------
    AssertionError
        If the dataframe does not have a column called 'phase_init'

    """

    assert 'phase_init' in df.columns, 'The dataframe must have a column called "phase_init"'
    df['dayN'] = df.phase_init.astype(int)
    df['dayN'] = [df.dayN.iloc[i] + 1 if df.phase_init.iloc[i] > 0 else df.dayN.iloc[i] for i in range(df.shape[0])]
    return df
```

#### Number of non detections in 5 days leading up to alert

Note the code below may seem hacky but it works. Here's the logic:
We've already counted the number of detections. The number of non detections is just the number of rows associated with each object minus the number of detections.

```python
NON_Nsince_min5d = lightcurves[lightcurves.dayN >= -5].groupby(['ATLAS_ID']).det.count() - DET_Nsince_min5d
```

## 5. "train-test" split and resampling

Now that we have our features and labels we need to split the data into train and test set, and we will also want to  do some resampling on the training set so that it is balanced - as we've seen the distribution of labels is very uneven.

### On the test set
* 1) The 'test' set here isn't a true test set as it's being looked at constantly in the training phase to see how well the models generalise. We will look at another - truer to the spirit - test set in another notebook. That one will be genuinely untouched until the models have been finalised and will provide a more genuine assessment of how well our VRA generalises. 
* 2) The **test sets throughout this project are unbalanced**. That's because we want our metrics to be calculated on data sets that are representative of what will be seen in the telescope pipeline. 
* 3) The split is **randon and no chronological**. We originally intended to do a chronological split between train and test as that can provide a better idea of how well a model will generalise with data drift. However the short baseline on which the data set was gathered means that the chronological split here had the test set falling on a particularly odd couple of weeks riddled with issues with the 03a ALTAS unit - we _want_ to train on these, and since we have an extra unseen test set that _is_ split chronologically from the current sample, we decided to just randomise the split here.

### Resampling

To balance the data set we do a simple subsampling of the garbage and proper motion (PM) classes. To try and keep as many samples as possible we chose to subsample down to the rough number of galactic objects in the training set (here 2200), which is slightly higher than the number of extra-galactic objects (1900), so technically there is a slight unbalance but we've not found this to be an issue.

```python
class SplitAndBalancedDataSetDay1():
    def __init__(self, X, y, vra_last_entries, train_fraction):
        self.X = X.join(vra_last_entries.mjd_init).sort_values('mjd_init').drop('mjd_init', axis=1)
        self.y = y.loc[self.X.index]
        self.train_fraction=train_fraction
        self.vra_last_entries=vra_last_entries
    

    def random_train_test_split(self):
        self.training_indexes = list(self.y.sample(int(self.X.shape[0] * self.train_fraction)).index)
        self.test_indexes = list(set(self.y.index.values)-set(self.training_indexes))
        
        self.X_train_unbalanced = self.X.loc[self.training_indexes]
        self.X_test_unbalanced = self.X.loc[self.test_indexes]
        self.y_train_unbalanced = self.y.loc[self.training_indexes]
        self.y_test_unbalanced = self.y.loc[self.test_indexes]      
        
    def subsample(self,N):
        self.N = N
        subsample_garbage = self.X_train_unbalanced[(self.y_train_unbalanced == 'garbage')
                              ].sample(self.N).index
        subsample_pm = self.X_train_unbalanced[(self.y_train_unbalanced == 'pm')
                              ].sample(self.N).index
        indexes_notgrabage = self.X_train_unbalanced[(self.y_train_unbalanced != 'garbage')
                                                     &(self.y_train_unbalanced != 'pm')
                                                    ].index

        indexes_training_sample = list(set(indexes_notgrabage).union(set(subsample_garbage)).union(set(subsample_pm)))

        self.X_train = self.X_train_unbalanced.loc[indexes_training_sample]
        self.y_train = self.y_train_unbalanced.loc[indexes_training_sample]
```
